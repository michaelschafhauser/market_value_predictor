## Creating list of lists of dfs

```python
all_dfs = []
for i in range(1992,2022):
    year_df = []
    for file_name in glob.glob(f"/home/alexpins/code/alexpins/Football project/transfer_history/{i}"+"/*.csv"):
        year_df.append(pd.read_csv(file_name[file_name.index("Football project/")+len("Football project/"):], low_memory=False))
    all_dfs.append(year_df)
```

## Concatenating all transfer history files from 1992

```python
years_dfs = []
for i in range(30):
    years_dfs.append(
        pd.concat([all_dfs[i][0], all_dfs[i][1], all_dfs[i][2], all_dfs[i][3],
                  all_dfs[i][4], all_dfs[i][5], all_dfs[i][6], all_dfs[i][7], all_dfs[i][8]]))
```

```python
transfer_history_combined = pd.concat([
    years_dfs[0], years_dfs[1], years_dfs[2], years_dfs[3], years_dfs[4],
    years_dfs[5], years_dfs[6], years_dfs[7], years_dfs[8], years_dfs[9],
    years_dfs[10], years_dfs[11], years_dfs[12], years_dfs[13], years_dfs[14],
    years_dfs[15], years_dfs[16], years_dfs[17], years_dfs[18], years_dfs[19],
    years_dfs[20], years_dfs[21], years_dfs[22], years_dfs[23], years_dfs[24],
    years_dfs[25], years_dfs[26], years_dfs[27], years_dfs[28], years_dfs[29]
])
```

```python
transfer_history_combined.reset_index(drop=True, inplace=True)
```

```python
transfer_history_combined.to_csv("transfer_history_combined.csv")
```
