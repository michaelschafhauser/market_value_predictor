```python
from market_value_predictor.predict import download_model, get_player_features
from forex_python.converter import CurrencyRates
import pandas as pd
import time
```

```python
df = pd.read_csv("../../raw_data/all_players_FUTDB_dump.csv")
```

```python
df = df.loc[(df["resource_id"]==df["resource_base_id"]) & (df["club"]!=112658)]
df.sort_values("rating", ascending=False).to_csv("../../raw_data/players_overview_total.csv")
```

```python
model = download_model()
```

```python
top_players = list(df.iloc[:1000].name)
```

```python
predictions = []
i = 1
for player in top_players:
    print(f"{i} - {player}")
    features, _ = get_player_features(player)
    
    if features.empty:
        predictions.append(None) 

    elif features.shape == (1, 1):
        predictions.append(None)

    else:
        prediction = model.predict(features)
        c = CurrencyRates()
        predictions.append(prediction[0] * float(c.get_rate('GBP', 'EUR')))
    i+=1
    if i==200 or i==400 or i==600 or i==800:
        print("Sleeping...")
        time.sleep(180)
```

```python
predictions
```

```python
pd.DataFrame({
    "name": top_players,
    "predicted_fee": predictions
}).dropna().sort_values("predicted_fee", ascending=False).reset_index(
    drop=True).to_csv("../../raw_data/top_players_prediction.csv")
```
