```python
import pandas as pd
```

```python
players_merged_by_name_df=pd.read_csv('../raw_data/players_merged_by_name_data')
players_merged_by_fifaid_df=pd.read_csv('../raw_data/player_merged_by_fifaid.csv')
```

```python
players_merged_by_name_df.drop('first_name', axis=1, inplace=True)
players_merged_by_name_df.drop('last_name', axis=1, inplace=True)
players_merged_by_name_df.drop('age_y', axis=1, inplace=True)
```

```python
players_merged_by_fifaid_df.drop('age_y', axis=1, inplace=True)
players_merged_by_fifaid_df.drop(columns=["Unnamed: 0.1"], inplace=True)
```

```python
concat_df = pd.concat([players_merged_by_fifaid_df,players_merged_by_name_df], axis="index", sort=False) 
```

```python
concat_df = concat_df.drop_duplicates(keep="first")
```
