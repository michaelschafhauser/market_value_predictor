```python
import pandas as pd

player_transfer_df = pd.read_csv('michaelschafhauser/market_value_predictor/raw_data/transfer_history_combined.csv') 
player_features_df = pd.read_csv('michaelschafhauser/market_value_predictor/raw_data/players_combined.csv') 
```

```python
player_transfer_df["cross_name"]=player_transfer_df["player_name"].str.extract(r'\b(\w+)$', expand=True)
player_transfer_df["cross_name"]=player_transfer_df["cross_name"].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
player_transfer_df.drop(player_transfer_df.filter(regex="Unname"),axis=1, inplace=True)
player_transfer_df["FIFA year"]=player_transfer_df["year"]+1
```

```python
player_features_df["cross_name"]=player_features_df["short_name"].str.extract(r'\b(\w+)$', expand=True)
player_features_df["cross_name"]=player_features_df["cross_name"].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
player_features_df.drop(player_features_df.filter(regex="Unname"),axis=1, inplace=True)
```

```python
merged_df = player_transfer_df.merge(player_features_df, on=["cross_name","FIFA year"], how='inner')
```
